```python
X=day.drop("cnt",axis=1)
y=day["cnt"]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,y_train,X_test,y_test=train_test_split(X,y,test_size=0.3,random_state=100)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train[["temp","hum","windspeed","casual","registered"]]=scaler.fit_transform(X_train[["temp","hum","windspeed","casual","registered"]])
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
model=GradientBoostingRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
